# Data Preprocessing and Data Cleaning

## Reading the data
To read the data from the Comma seperated value file, we need to import the file, read and then display the details of the dataset.

```python
#Import Statements
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Import the dataset
data = pd.read_excel('GSAF5-1.xls')
pd.set_option('display.max_columns', 999)
```

```python
#Displaying all the columns in the dataset
data.columns
```

## Data Cleaning
Data Cleaning is the first and foremost step in handling a dataset. This step is more important because the quality of the dataset is acquired here. 

In the process of cleaning the dataset, we will encounter some missing values, categorical values, unwanted information, etc., We need to overcome these obstacles inorder to get a better quality dataset.

```python
#Have renamed the column 'Sex' by eliminating the space at the end of the column
data.rename(columns ={'Sex ':'Sex'},inplace=True)
data
```

```python
#Deleted the column 'Unnanmed: 23' as it was empty
del(data['Unnamed: 23'])

#Deleted the column 'Unnanmed: 22' as it was empty
del(data['Unnamed: 22'])
```

```python
#Removed the empty cells after the cell 6451
data = data.drop(data.index[6451:])
data
```

```python
#Creating a new dataset which contains only the data related to USA
data = data[data.Country == "USA"]
data
```

```python
#Showing the data description
data.describe()
```

```python
data.info()
```

```python
data.head()
```

```python
#Checking whether any data contains NULL value and printing the sum of the null values
data.isnull().sum()
```

```python
#Replacing all null values at sex column with M (Male)
data["Sex"].fillna("M", inplace = True)
```

```python
# Finding unique values from Sex column
data['Sex'].unique()
```

```python
# Cleaning the Sex Column by replacing the missing values and unrelated values by "M"
data[data['Sex'] == 'M '] = 'M'
data[data['Sex'] == 'M x 2'] = 'M'
data[data["Sex"]=='lli'] = 'M'
data['Sex'].unique()
```

```python
data['Age'] = pd.to_numeric(data['Age'],errors='coerce')
```

```python
#Calculating mean value of the age column
data['Age'].mean()
```

```python
#Replacing all null values at age column with 27 (Mean value of the age column)
data['Age'] = data['Age'].fillna(27)
data['Age']
```

```python
#Removing row with age = young
data = data[data.Age != 'young']
```

# Data Visualization

```python
#Plotting the Sex column in Countplot for knowing the Male and Female dispersement
sns.countplot(x='Sex',data=data)
```

# Output-1
Plotting age distribution through binning of the age column

```python
#Age distribution graph
fig,ax = plt.subplots(figsize=(8,6))
ax = sns.distplot(data['Age'],
             ax=ax,
             hist_kws={"alpha": 0.6, "color": "#00c2c7"},
             kde=False,bins=15)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distribution')
```

```python
# Grouping the data based on Age and Area columns
age_data = pd.DataFrame(data['Age'].groupby(data['Area']).mean())
age_data.Age
```

```python
#Importing data set-2 containing state and region
data_set = pd.read_csv('data_2.csv')
data_set
```

```python
#Merging two datasets 
merge_data = pd.merge(left=data, right=data_set, on=None, left_on='Area', right_on='State')
merge_data
```

```python
#Displaying the columns in the second dataset
merge_data.columns
```

```python
#Displaying the Gender distribution based on the region
region_gender = merge_data.groupby(['Sex','Region']).size().to_frame('count').reset_index()
region_gender
```

```python
pivot_df_region = region_gender.pivot(index='Region', columns='Sex', values='count')
```

# Output-4
